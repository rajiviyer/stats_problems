5. A company manufactures LED bulbs with a faulty rate of 30%. If I randomly select 6
chosen LEDs, what is the probability of having 2 faulty LEDs in my sample?
Calculate the average value of this process. Also evaluate the standard deviation
associated with it.

```python
import scipy.stats as stats
fault_rate = 0.30
led_count = 6
faulty_count = 2
print(f"Probability of having 2 faulty LEDs: {np.round(stats.binom.pmf(k=faulty_count,n=led_count, p=fault_rate),2)}")
process_mean,process_var = stats.binom.stats(n=led_count,p=fault_rate)
print(f"Mean of Process: {np.round(process_mean,2)}")
print(f"Standard Deviation of Process: {np.round(np.sqrt(process_var),2)}")
```

6. Gaurav and Barakha are both preparing for entrance exams. Gaurav attempts to
solve 8 questions per day with a correction rate of 75%, while Barakha averages
around 12 questions per day with a correction rate of 45%. What is the probability

factors affecting their ability to solve questions correctly? Give a pictorial

7. Customers arrive at a rate of 72 per hour to my shop. What is the probability of k
customers arriving in 4 minutes? a) 5 customers, b) not more than 3 customers, c)
more than 3 customers. Give a pictorial representation of the same to validate your
answer.

* **Solution** This is a poisson distribution $P(x; \mu) = \frac {(e^{-\mu}) (\mu ^{x})} {x!}$ \\
a) 72 customers per hour = 1.2 customers per min = 4.8 customers every 4 mins. Hence $\mu=1.2$. This is a poisson distribution. Hence Probability of k customers arriving in 4 minutes : $P(x; \mu) = \frac {(e^{-\mu}) (\mu ^{x})} {x!}=\frac {(e^{-4.8}) (4.8 ^{k})} {k!}$

```python
#b) 5 customers
print(f"Probability of 5 customers arriving in 4 mins is :{np.round(stats.poisson.pmf(k=5, mu=4.8),2)}")
```

```python
#c not more than 3 customers
print(f"Probability of not more than 3 customers arriving in 4 mins (i.e 3 customers or less) is :{np.round(stats.poisson.cdf(k=3, mu=4.8),2)}")
print(f"Probability of more than 3 customers arriving in 4 mins is :{np.round(stats.poisson.sf(k=3, mu=4.8),2)}")
```

8. I work as a data analyst in Aeon Learning Pvt. Ltd. After analyzing data, I make
reports, where I have the efficiency of entering 77 words per minute with 6 errors per
hour. What is the probability that I will commit 2 errors in a 455-word financial report?
What happens when the no. of words increases/decreases (in case of 1000 words,
255 words)?
How is the λ affected?
How does it influence the PMF?
Give a pictorial representation of the same to validate your answer.

```python
#Time taken to type 455 words
time_455 = np.round(455./77.,3)
mean_error_per_minute = np.round(6./60.,2)
print(f"Time taken for 455 page report :{mean_error_per_minute} minutes")
print(f"Mean error for 455 page report :{np.round(mean_error_per_minute*time_455,3)}")
print(f"Probability for 2 errors in 455 page report :{np.round(stats.poisson.pmf(k=2, mu=mean_error_per_minute*time_455),2)}")
print("-------------")
time_1000 = np.round(1000./77.,3)
mean_error_per_minute = np.round(6./60.,2)
print(f"Time taken for 1000 page report :{mean_error_per_minute} minutes")
print(f"Mean error for 1000 page report :{np.round(mean_error_per_minute*time_1000,3)}")
print(f"Probability for 2 errors in 1000 page report :{np.round(stats.poisson.pmf(k=2, mu=mean_error_per_minute*time_1000),2)}")
print("-------------")
time_255 = np.round(255./77.,3)
mean_error_per_minute = np.round(6./60.,2)
print(f"Time taken for 255 page report :{mean_error_per_minute} minutes")
print(f"Mean error for 255 page report :{np.round(mean_error_per_minute*time_255,3)}")
print(f"Probability for 2 errors in 255 page report :{np.round(stats.poisson.pmf(k=2, mu=mean_error_per_minute*time_255),2)}")
```

> As expected, mean errors $\lambda$ is directly proportional to the number of words to be typed and probability (PMF) of 2 errors occurring will increase as the number of pages to be typed increase

11. The current flow in a copper wire follow a normal distribution with a mean of 10 mA
and a variance of 4 (mA)2.
What is the probability that a current measurement will exceed 13 mA? What is the
probability that a current measurement is between 9 and 11mA? Determine the
current measurement which has a probability of 0.98.

```python
mean_flow=10
var_flow=4
sd_flow = np.sqrt(var_flow)
print(f"The probability that the current measurement will exceed 13ma :{np.round(stats.norm.sf(13,loc=mean_flow, scale=sd_flow),3)}")
print(f"The probability that the current measurement is between 9 and 11 ma is :{np.round(stats.norm.cdf(11,loc=mean_flow, scale=sd_flow) - stats.norm.cdf(9,loc=mean_flow, scale=sd_flow),3)}")
print(f"The current measurement having a probability of 0.98 is :{np.round(stats.norm.ppf(0.98,loc=mean_flow, scale=sd_flow),3)} ma")
```
